```python
import nltk
from nltk.stem.lancaster import LancasterStemmer
nltk.download('punkt')
import os
import json
import datetime
stemmer = LancasterStemmer()
```

```python
training_data = []
training_data.append({"class":"greeting", "sentence":"how are you?"})
training_data.append({"class":"greeting", "sentence":"Good Morning"})
training_data.append({"class":"greeting", "sentence":"Whats up"})

training_data.append({"class":"introduction", "sentence":"What is your name?"})
training_data.append({"class":"introduction", "sentence":"What do you study?"})
training_data.append({"class":"introduction", "sentence":"Where do you live"})

training_data.append({"class":"weather", "sentence":"How is the weather today?"})
training_data.append({"class":"weather", "sentence":"Is it raining?"})
training_data.append({"class":"weather", "sentence":"What is the temperature outside?"})

print("%s sentences in training data" % len(training_data))
```

```python
doc_words = []
documents = []
classes = []
ignore_words = ['?']

for pattern in training_data:
    w = nltk.word_tokenize(pattern['sentence'])
    #print("w is %s" % w)
    doc_words.extend(w)
    print("words is %s" % doc_words)
    documents.append((w,pattern['class']))
    #print("documents is %s" % documents)
    
    if pattern['class'] not in classes:
        classes.append((pattern['class']))
        #print("Class is %s" % classes)

print(len(doc_words))
print(set(classes))
```

```python
words = [stemmer.stem(w.lower()) for w in doc_words if w not in ignore_words]
words = list(set(words))

classes = list(set(classes))

print(len(documents), "documents")
print(len(words), "unique stemmed words", words)
print(len(classes), "classes", classes)
```

```python
training_data = []
output_data = []
output_empty = [0] * len(classes)


for doc in documents:
    bag = []
    pattern_words = doc[0]
    print(pattern_words)
    pattern_words = [stemmer.stem(word.lower()) for word in pattern_words]
    print(pattern_words)
    for w in words:
        print(w)
        bag.append(1) if w in pattern_words else bag.append(0)
        print(bag)
    
    training_data.append(bag)
    print(training_data)
    output_row = list(output_empty)
    output_row[classes.index(doc[1])] = 1
    print(classes.index(doc[1]))
    output_data.append(output_row)
    print(output_data)

print(training_data)
print(output_data)
#i=0
#w = documents[i][0]
#print(w)
#print ([stemmer.stem(word.lower()) for word in w])
#print (training_data[i])
#print (output_data[i])
    
```
